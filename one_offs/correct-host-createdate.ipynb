```python
from pymongo import MongoClient

mongo_client = MongoClient('')
db = mongo_client['phishstory']
collection = db['incidents']
```

```python
from dateutil import parser
import pymongo

results = collection.find({ '$and': [ {'data.domainQuery.host.createdDate': {'$ne': 'None'}}, {'data.domainQuery.host.createdDate': {'$type': 'string'}}, {'failedEnrichment': { '$exists': True }} ] })
i = 0
for ticket in results:
    try:
        new_date = parser.parse(ticket.get('data').get('domainQuery').get('host').get('createdDate'))
        updates = { 'data.domainQuery.host.createdDate': new_date}
        document = collection.find_one_and_update({'_id': ticket["_id"]}, {'$set': updates}, upsert=False, return_document=pymongo.ReturnDocument.AFTER)
        print(f'{ticket["_id"]} Updated - {updates}')
    except Exception as e:
        print(f'{ticket["_id"]} - update failed {e}')
    i += 1
print(i)
```
